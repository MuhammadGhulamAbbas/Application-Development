##  Pandas loc[ ] , iloc[ ] and ix[ ]  based Data Retrieval 

Aprat from using index operator( [] ) , Pandas provides some other ways like loc[] and  iloc[ ]  for efficient data retrieval. Collectively these are called Indexers.

In order to understand the use of loc [] and iloc[] we need to understand Pandas indexing mechanism.

Pandas provides a default index with each DataFrame which is an integer index. Incase of default integer index loc[] provides all the functionality of iloc[] . Apart from using Pandas default index , we can define our own index which can be a non-integer type like Date , Object etc

### Remember :

1. loc[] is label-based, which means that you have to specify rows and columns based on their row and column labels.
2. iloc[] is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).

#### loc[] Demonstration

```python
# Access all rows across age and marital
df.loc[:,['education','marital']]
```
