# 1. Working with Files 

###  Why we need Files ?
1. Programs we have written so far , use console input / output
2. All values sotred in variables are stored in RAM
3. Output on console is also stored in RAM 
4. RAM is volatile. When you close the program or application , all data ( input /output) is lost


###  How files are useful ?
When a program needs to save data for later use, it writes the data in a
file. The data can be read from the file at a later time.

### Types of Files 
#### 1. Text File
   A text file in Python is a file that contains only text characters. Text files are typically encoded in a standard format, such as ASCII or UTF-8, which allows them to be read and written by a variety of software applications.

Text files can be used to store a variety of data, such as plain text, source code, configuration files, and log files. Text files are also often used to store data that is exchanged between different software applications.
   
As a result, the file may be opened and viewed in a text editor such as Notepad. Examples : txt , csv , py files etc
#### 2. Binary Files 
  A binary file in Python is a file that contains any type of data, including text, images, audio, video, and executable programs. Binary files are stored in a binary format, which means that they are made up of a series of 0s and 1s.
  
  We will be dealing with text files here

### File Access Methods 
Most programming languages provide two different ways to access data stored in a file. 
1. Sequential access ( Data is read from the beginning of the file to the end . Media on cassette tape )
2. Random or Direct access. ( Data can be read directly from any point. Media on Hard Disk etc)

We will be dealing with sequential access files 

### File Operations 
There are always
three steps that must be taken when a file is used by a program.
1. Open the file—Opening a file creates a connection between the file and the program.
Opening an output file usually creates the file on the disk and allows the program to
write data to it. Opening an input file allows the program to read data from the file.
2. Process the file—In this step data is either written to the file (if it is an output file) or
read from the file (if it is an input file).
3. Close the file—When the program is finished using the file, the file must be closed.
Closing a file disconnects the file from the program.

## Working with Files

### Opening a File 

1. In order for a program to work with a file on the computer’s disk, the program must create
a file object in memory.
2. A file object is an object that is associated with a specific file and provides a way for the program to work with that file.
3. In the program, a variable references the file object. This variable is used to carry out any operations that are performed on the file.

 You use the open function in Python to open a file. The open function creates a file object
 and associates it with a file on the disk. Here is the general format of how the open function
 is used:

file_variable = open(filename, mode)



### Writing to a File

```python
myfile = open('test.txt', 'w')
myfile.write('My first file written from Python\n')
myfile.write('---------------------------------\n')
myfile.write('Hello, world!')
myfile.close()
```

The first line opens the test.txt for writing. If there is no file named test.txt on the disk, it will be created. If there already is one, it will be replaced by the file we are writing and any previous data in it will be lost.

Closing the file handle tells the system that we are done writing and makes the disk file available for reading by other programs (or by our own program).

### Counting Number of Senetnces and Words in a Text File

```python
filename = open('sample.txt', 'r')
```

```python
def count_sentences_and_words(filename):
    f=open(filename, "r") 
    text=f.read()
    # Split the text into sentences.
    sentences = text.split(".")
    # Split the text into words.
    words = text.split()
    return len(sentences), len(words)


# Example usage:

filename = "sample.txt"
num_sentences, num_words = count_sentences_and_words(filename)

print(f"Number of sentences: {num_sentences}")
print(f"Number of words: {num_words}")
```

### Using loops to enter data into files 

```python
def main():
    #Get the number of days.
    num_days = int(input('For how many days do you have sales? '))
    # Open a new file named sales.txt.
    sales_file = open('sales.txt', 'w')

    # Get the amount of sales for each day and write it to the file.
    for count in range(1, num_days + 1):
    #Get the sales for a day.
        sales = float(input('Enter the sales for day # '+ str(count) + ': '))

        # Write the sales amount to the file.
        sales_file.write(str(sales) + '\n')
        
    # Close the file.
    sales_file.close()
    print('Data written to sales.txt.')

#Call the main function.
main()
```

### Using loops to read data from files 

```python
def main():
    
    sales_file = open('sales.txt', 'r')
    line = sales_file.readline()

    while line != '':
        amount = float(line)
        
        print(format(amount, '.2f'))
        line = sales_file.readline()

    sales_file.close()
    
main()
```

# 2. Exception Handling

### What is an Exception?
An exception is an error that occurs while a program is running, causing the program to abruptly halt. 

```python
# This program divides a number by another number.

def main():
    # Get two numbers.
    num1 = int(input('Enter a number: '))
    num2 = int(input('Enter another number: '))

    # Divide num1 by num2 and display the result.
    result = num1 / num2
    print(num1, 'divided by', num2, 'is', result)

# Call the main function.
main()
```

### What is traceback?
The lengthy error message that is shown when a run time error occurs is called a traceback. The traceback gives information regarding the line number(s) that caused the exception. 

### How exceptions can be handled ?
We can use try / except mechanism to handle exceptions .
1. Any statement or group of statements that can potentially rasie an exception are included in try block.
2. If an exception occures in try block , all the statements following the statement that caused the exception are skipped and control is transferred to the except block .
3. In except block , a proper message regarding the exception is displayed and program is terminated gracefully

```python
#  one simple way to avoid exceptions is to code your program carefully 
def main():
    # Get two numbers.
    num1 = int(input('Enter a number: '))
    num2 = int(input('Enter another number: '))

    # Divide num1 by num2 and display the result.
    if(num2!=0):
        result = num1 / num2
        print(num1, 'divided by', num2, 'is', result)
    else:
        print("Denominator cannot be zero")
# Call the main function.
main()
```

```python
# This program divides a number by another number.

def main():
    # Get two numbers.
    try:
        num1 = int(input('Enter a number: '))
        num2 = int(input('Enter another number: '))

        # Divide num1 by num2 and display the result.
        result = num1 / num2
        print(num1, 'divided by', num2, 'is', result)
    except ArithmeticError as r :
        print("The following error has occured:",r )
# Call the main function.
main()
```

### Handling multiple exceptions 

```python

# This program displays the total of the amounts in the sales.txt file.
def main():
    # Initialize an accumulator.
    total = 0.0
    try:
        # Open the sales_data.txt file.
        infile = open('sales.txt', 'r')

        # Read the values from the file and # accumulate them.
        for line in infile:
            amount = float(line)
            total += amount

        # Close the file.
        infile.close()

        # Print the total.
        print(format(total, ',.2f'))
    except IOError:
        print('An error occured trying to read the file.')
    except ValueError:
        print('Non-numeric data found in the file.')

    except:
        print('An error occured.')
# Call the main function.
main()
```
